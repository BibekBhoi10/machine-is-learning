# GRADIENT DESCENT AND COST FUNCTION

#### cost function :
   The cost function measures the difference between the predicted output and actual values.
   ##### common type :
     MSE( Mean Squared Error) 
#### gradient descent: 
   it is an algorithm to minimize the cost function by updating the model parameters and also to find the best fit line for given training dataset

## $$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - y_{\text{predicted}})^2 $$
### $$ y_{\text{predicted}} = m x_i + b $$
Where :
- n = Number of data points

- yi = Actual values

- ypredicted = predicted values

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python

def gradient_descent (x,y):
    m_curr = b_curr = 0
    learning_rate = 0.01
    n = len(x)
    for i in range(1000):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)]) # cost function
        # m derivative 
        md = -(2/n)*sum(x*(y-y_predicted))
        # b derivative 
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd  
        print(f"cost = {cost}, m = {m_curr}, b = {b_curr}")
```

#### new_m = current_m - learning_rate * md 
#### similarly new_b = current_b - learning_rate * bd

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

### plotting the graph

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0  # Initialize parameters
    rate = 0.01  # Learning rate
    n = len(x)

    plt.scatter(x, y, color='red', marker='+', linewidth=5)  # Plot data points

    for i in range(1000):  # 10,000 is too high for visualization
        y_predicted = m_curr * x + b_curr
        
        # Compute gradients
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
        # Update parameters
        m_curr -= rate * md
        b_curr -= rate * bd
        
        #plot the line
        plt.plot(x, y_predicted, color='green')
    
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Gradient Descent Visualization")
    plt.show()
    
    return m_curr, b_curr  # Return final values

# Sample Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Run the function
final_m, final_b = gradient_descent(x, y)
print(f"Final m: {final_m}, Finalb:{final_b}")
```
