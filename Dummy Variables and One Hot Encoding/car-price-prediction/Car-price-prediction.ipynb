# Car Selling Price Prediction Model

### Exercise
At the same level as this notebook on github, there is an Exercise folder that contains carprices.csv. This file has car sell prices for 3 different models. First plot data points on a scatter plot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions,

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score()

```python
import pandas as pd
df = pd.read_csv('carprices.csv')
df
```

## plotting the graph 

```python
import matplotlib.pyplot as plt

# Mileage vs Sell Price($)
plt.scatter(df[['Sell Price($)']], df['Mileage'], color = 'red')
plt.xlabel('Sell Price($)')
plt.ylabel('Mileage')
plt.title('Mileage vs Sell Price($)')
plt.show()

# Age(yrs) vs Sell Price($)
plt.scatter(df[['Sell Price($)']], df['Age(yrs)'])
plt.xlabel('Sell Price($)')
plt.ylabel('Age(yrs)')
plt.title('Age(yrs) vs Sell Price($)')
plt.show()
```

```python
dummies = pd.get_dummies(df['Car Model']).astype(int)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
# now dropping columns car model and one of the dummy variables
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis = 'columns')
final
```

```python
# to get X dataset drop the sell price columns
X = final.drop(['Sell Price($)'],axis = 'columns')
X
```

```python
y = final['Sell Price($)']
y
```

### now applying linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

**Price of mercedez benz that is 4 yr old with mileage 45000**

```python
model.predict([[45000,4,0,0]])
```

**Price of BMW X5 that is 7 yr old with mileage 86000**

```python
model.predict([[86000,7,0,1]])
```

### saving the model 

```python
import joblib
```

```python
joblib.dump(model, 'car_price_predict_model')
```

```python
car_model = joblib.load('car_price_predict_model')
```

```python
car_model.predict([[86000,7,0,1]])
```
