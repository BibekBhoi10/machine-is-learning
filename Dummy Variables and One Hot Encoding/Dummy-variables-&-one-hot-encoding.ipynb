# DUMMY VARIABLES AND ONE HOT ENCODING

### For dealing with strings or words i.e categorical varibles one hot encoding method and dummy variable is used

##### categorical variables are of two types :
- Nominal : Non numeric such as (Male female), (green red blue) , etc
- Ordinal : Numeric variable such (high medium low) , (graduate masters phd) , ( satisfied neutral dissatisfied)

```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices.csv')
df
```

## using pandas to create dummy variables

```python
dummies = pd.get_dummies(df['town']).astype(int)
dummies
```

```python
merged = pd.concat([df,dummies] , axis = 'columns')
merged
```

## droping the columns

#### droping the town column and also one of the dummy variable column to avoid dummy variable trap 
##### dummy variable trap :
The Dummy Variable Trap is a situation that occurs when you include all dummy variables in a regression model, leading to multicollinearity (i.e., high correlation between independent variables). This can cause issues in models like linear regression where predictors should not be highly correlated.

```python
final = merged.drop(['town','west windsor'], axis = 'columns')
final
```

## now implementing the linear regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = final.drop(['price'], axis = 'columns')
X
```

```python
y = df['price']
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

In Scikit-Learn, the .score(X, y) method evaluates the performance of a trained model on a given dataset (X, y)

```python
model.score(X,y)
```

## One Hot Encoding

First step is to use label encoder to convert town names into numbers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town) # fit and transform convert the town column to label
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
y = dfle['price']
y
```

## creating dummy variables

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[:,1:] # X[take all the rows, take all the columns after 1]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[3400,0,0]])
```

both prediction using dummy variables and One Hot Encoding gives the same result
