# K fold cross validation 

#### K-Fold Cross Validation is a technique used in machine learning to evaluate how well a model performs on unseen data. It helps in reducing overfitting and gives a more accurate estimate of model performance

## How it Works:
Split the dataset into K equal parts (folds).

For each fold:

- Use that fold as the validation set.
- Use the remaining K-1 folds as the training set.
- Train and evaluate the model K times.
- Average the results (accuracy, error, etc.) across all K runs.

```python

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier
import numpy as np 
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
X = digits.data
X
```

```python
y = digits.target
y
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

**Logistic Regression**

```python
lr = LogisticRegression(max_iter=500)
lr.fit(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```

**SVM**

```python
svm = SVC(gamma='auto')
svm.fit(X_train, y_train)
```

```python
svm.score(X_test, y_test)
```

**Random Forest**

```python
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

# K Fold cross Validation

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 3)
kf
```

```python
## for basic example
for train_index, test_index in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_index, test_index)
```

```python
def get_score(model , X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)
```

```python
scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    scores_logistic.append(get_score(LogisticRegression(solver='liblinear'), X_train, X_test, y_train, y_test))  
    scores_svm.append(get_score(SVC(gamma='auto'), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
```

```python
scores_logistic
```

```python
scores_svm
```

```python
scores_rf
```

## cross_val_score function

```python
from sklearn.model_selection import cross_val_score
```

**Logistic regression model performance using cross_val_score**

```python
cross_val_score(LogisticRegression(max_iter = 500), digits.data, digits.target)
```

**svm model performance using cross_val_score**

```python
cross_val_score(SVC(), digits.data, digits.target)
```

**random forest performance using cross_val_score**

```python
cross_val_score(RandomForestClassifier(),digits.data, digits.target)
```

## Parameter Tuning using K Fold

```python
scores1 = cross_val_score(RandomForestClassifier(n_estimators=5),digits.data, digits.target, cv=10)
np.average(scores1)
```

```python
scores2 = cross_val_score(RandomForestClassifier(n_estimators=20),digits.data, digits.target, cv=10)
np.average(scores2)
```

```python
scores3 = cross_val_score(RandomForestClassifier(n_estimators=30),digits.data, digits.target, cv=10)
np.average(scores3)
```

```python
scores4 = cross_val_score(RandomForestClassifier(n_estimators=40),digits.data, digits.target, cv=10)
np.average(scores4)
```

Here we used cross_val_score to fine tune our random forest classifier and figured that having around 30 trees in random forest gives best result.
