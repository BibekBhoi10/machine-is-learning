# LINEAR REGRESSION

Linear Regression is a supervised machine learning algorithm used for predicting numerical values based on given input data. It finds a straight-line relationship between independent variables (X) and dependent variables (Y).

Mathematical Equation of Linear Regression

Y = mX+c

Where:

Y = Predicted value (dependent variable)

X = Input feature (independent variable)

m = Slope of the line (how much Y changes when X changes)

c = Intercept (value of Y when X = 0)

# Import dependencies

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# import data

```python
df = pd.read_csv("houseprices.csv")
df
```

# ploting the data in the graph

```python
%matplotlib inline
plt.xlabel('prices')
plt.ylabel('area')
plt.scatter(df.area,df.price, color = "red", marker = "+")
```

# making of linear regression model

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["area"]],df.price)
```

# now lets start prediction

```python
reg.predict([[3100]])
```

```python
# to get the coefficient
reg.coef_
```

```python
# to get the intercept
reg.intercept_
```

## y = mx+b 
y = prices, 
m = slope, 
x = area, 
b = intercept

```python
d = pd.read_csv("areas.csv")
d
```

```python
p=reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.to_csv("prediction.csv", index = False)
```

# plotting the prediction line 

```python
%matplotlib inline
plt.xlabel('prices')
plt.ylabel('area')
plt.scatter(df.area,df.price, color = "red", marker = "+")
plt.plot(df.area, reg.predict(df[['area']]),color = 'blue')
```
